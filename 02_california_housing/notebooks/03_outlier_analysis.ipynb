## target distribution

```python
import sys
import os

sys.path.append(os.path.abspath(".."))

from src.data_loader import load_data

import matplotlib.pyplot as plt
import seaborn as sns

df = load_data()

plt.figure(figsize=(8,5))
sns.histplot(df["MedHouseVal"], bins=50)

plt.title("Target Distribution")
plt.show()
```

## target outlier detection

```python
plt.figure(figsize=(8,3))

sns.boxplot(
    x=df["MedHouseVal"]
)

plt.title("Target Boxplot")
plt.show()
```

## log transform target

```python
import numpy as np

df["MedHouseVal_log"] = np.log1p(
    df["MedHouseVal"]
)

plt.figure(figsize=(8,5))

sns.histplot(
    df["MedHouseVal_log"],
    bins=50
)

plt.title("Log Transformed Target")
plt.show()
```

## feature outlier analysis

```python
numeric_features = [
    "MedInc",
    "AveRooms",
    "AveBedrms",
    "Population",
    "AveOccup"
]

for col in numeric_features:

    plt.figure(figsize=(8,3))

    sns.boxplot(
        x=df[col]
    )

    plt.title(col)

    plt.show()
```
